```python
salestest = sales[(sales.month>11)][#&(sales.month<34)
    featurestscommon+otherfeaturescommon+categoricalfeaturescommon+
    featurestsfull+otherfeaturesfull+categoricalfeaturesfull+['target']
]
del sales
```

```python
featuresscaled = featurestscommon+otherfeaturescommon+featurestsfull+otherfeaturesfull

scalerfeatures = MinMaxScaler(feature_range=(0, 1))
scalerfeatures.fit(salestest[featuresscaled])

scalertarget = MinMaxScaler(feature_range=(0, 1))
scalertarget.fit(salestest[salestest.month < 34]['target'].values.reshape(-1,1))

salesscaled = pd.DataFrame(data=scalerfeatures.transform(salestest[featuresscaled]),columns=featuresscaled,dtype=np.float16)
salesscaled['target'] = pd.DataFrame(data=scalertarget.transform(salestest['target'].values.reshape(-1,1)),columns=['target'],dtype=np.float16)['target']

with pd.option_context('display.max_rows', 200, 'display.max_columns', 200):
    display(salesscaled.head(3))
```

```python
categoricalfeatures = categoricalfeaturescommon + categoricalfeaturesfull

salesscaled['monthnotscaled'] = salestest['month'].values
salesscaled['monthnotscaled'] = salesscaled['monthnotscaled'].astype(np.int8)
salesscaled['shopidnotscaled'] = salestest['shopid'].values
salesscaled['shopidnotscaled'] = salesscaled['shopidnotscaled'].astype(np.int8)
salesscaled['itemidnotscaled'] = salestest['itemid'].values
for categoricalfeature in categoricalfeatures:
    salesscaled[categoricalfeature] = salestest[categoricalfeature].values
    if categoricalfeature != 'itemid':
        salesscaled[categoricalfeature] = salesscaled[categoricalfeature].astype(np.int8)
    else:
        print('no int8')
    
del salestest
```

```python
def consecutive_categorical_ids(feature):
    print(feature)
    max_id = np.max(salesscaled[feature].unique())
    len_id = len(salesscaled[feature].unique())
    if max_id == len_id - 1:
        print('OK')
    else:
        print('NOK, max: {}, len: {}'.format(max_id,len_id))
```

```python
for categoricalfeature in categoricalfeatures:
    encoder = LabelEncoder()
    salesscaled[categoricalfeature] = encoder.fit_transform(salesscaled[categoricalfeature])
    if categoricalfeature != 'itemid':
        salesscaled[categoricalfeature] = salesscaled[categoricalfeature].astype(np.int8)
    else:
        print('no int8')
    consecutive_categorical_ids(categoricalfeature)
```

```python
tsinputscommon = Input(shape=(12,24))

#conv = Conv2D(64,(3,24),(2,24),padding='same')(tsinputscommon)
#convreshape = Reshape((6,64))(conv)

gru1common = GRU(64,return_sequences=True)(tsinputscommon)
gru2common = GRU(64)(gru1common)
#gru3common = GRU(128)(gru1common)

#tsinputsfull= Input(shape=(12,2))
#gru1full = GRU(16,return_sequences=True)(tsinputsfull)
#gru2full = GRU(16)(gru1full)


shopidinputs = Input(shape=(1,))
shopidembedding = Embedding(
    output_dim=6,
    input_dim=shopidnb,
    input_length=1)(shopidinputs)
shopidembedding = Reshape((6,))(shopidembedding)

shoptypeidinputs = Input(shape=(1,))
shoptypeidembedding = Embedding(
    output_dim=2,
    input_dim=shoptypeidnb,
    input_length=1)(shoptypeidinputs)
shoptypeidembedding = Reshape((2,))(shoptypeidembedding)

shopcityididinputs = Input(shape=(1,))
shopcityidembedding = Embedding(
    output_dim=4,
    input_dim=shopcityidnb,
    input_length=1)(shopcityididinputs)
shopcityidembedding = Reshape((4,))(shopcityidembedding)

itemcategoryidinputs = Input(shape=(1,))
itemcategoryidembedding = Embedding(
    output_dim=12,
    input_dim=itemcategoryidnb,
    input_length=1)(itemcategoryidinputs)
itemcategoryidembedding = Reshape((12,))(itemcategoryidembedding)

itemcategorytypeidinputs = Input(shape=(1,))
itemcategorytypeidembedding = Embedding(
    output_dim=3,
    input_dim=itemcategorytypeidnb,
    input_length=1)(itemcategorytypeidinputs)
itemcategorytypeidembedding = Reshape((3,))(itemcategorytypeidembedding)

itemcategorysubtypeidinputs = Input(shape=(1,))
itemcategorysubtypeidembedding = Embedding(
    output_dim=6,
    input_dim=itemcategorysubtypeidnb,
    input_length=1)(itemcategorysubtypeidinputs)
itemcategorysubtypeidembedding = Reshape((6,))(itemcategorysubtypeidembedding)

#itemidinputs = Input(shape=(1,))
#itemidembedding = Embedding(
#    output_dim=10,
#    input_dim=itemidnb,
#    input_length=1)(itemidinputs)
#itemidembedding = Reshape((10,))(itemidembedding)


otherinputscommon = Input(shape=(7,))

otherinputsfull = Input(shape=(4,))

concatenated_layer = Concatenate()([
    gru2common,
    #gru2full,
    shopidembedding,
    shoptypeidembedding,
    shopcityidembedding,
    itemcategoryidembedding,
    itemcategorytypeidembedding,
    itemcategorysubtypeidembedding,
    #itemidembedding,
    otherinputscommon,
    otherinputsfull
])

dense1 = Dense(128)(concatenated_layer)

outs = Dense(1)(dense1)

model = Model(inputs=[
    tsinputscommon,
    #tsinputsfull,
    shopidinputs,
    shoptypeidinputs,
    shopcityididinputs,
    itemcategoryidinputs,
    itemcategorytypeidinputs,
    itemcategorysubtypeidinputs,
    #itemidinputs,
    otherinputscommon,
    otherinputsfull
],outputs=outs)
model.compile(optimizer='adam', loss='mse')
```
